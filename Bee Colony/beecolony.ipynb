```python
import pandas as pd
import numpy as np
import random
```

```python
#f1 |xsin(x)/(2x-5)| where 0<=x<=14
#f2 e^(x^3-x) where -1<=x<=0
#f3 -(3*y)/(x^2+y^2+1) where -5<=x<=5 and -5<=y<=5
#f4 3(1-x)^2e^-x^2-(y+1)^2-10e^(-x^2-y^2)(-x^3+x/5-y^5)-1/3e^-(x+1)^2-y^2 where -3<=x<=3 and -3<=y<=3



#Create a bee colony algorithm to find the maximum of the functions

def f1(x):
    if 2*x-5 < 2:
        return 0
    return np.absolute(x*np.sin(x)/(2*x-5))
def f2(x):
    return np.exp(x**3-x)
def f3(x,y):
    return -(3*y)/(x**2+y**2+1)
def f4(x,y):
    return 3*(1-x)**2*np.exp(-x**2)-(y+1)**2-10*np.exp(-x**2-y**2)*(-x**3+x/5-y**5)-1/3*np.exp(-(x+1)**2-y**2)

def fitness1(x):
    return f1(x)
def fitness2(x):
    return f2(x)
def fitness3(x,y):
    return f3(x,y)
def fitness4(x,y):
    return f4(x,y)

#The bee algorithm has 3 types of bees: employed, onlooker, and scout bees 
#Employed bees search for food sources in the neighborhood of the food source they are currently exploiting
#Onlooker bees watch the dances of employed bees and choose a food source depending on the dances
#Scout bees search for new food sources when the employed bees have exhausted the food sources in their neighborhood

#The algorithm has 3 main steps: initialization, employed bee phase, and onlooker bee phase
#Initialization: Randomly generate a population of food sources
#Employed bee phase: Employed bees search for food sources in their neighborhood
#Onlooker bee phase: Onlooker bees watch the dances of employed bees and choose a food source depending on the dances

#Initialize the population of 
def initialize_population1(n):
    population = []
    for i in range(n):
        x = random.uniform(0,14)
        population.append(x)
    return population

def initialize_population2(n):
    population = []
    for i in range(n):
        x = random.uniform(-1,0)
        population.append(x)
    return population

def initialize_population3(n):
    population = []
    for i in range(n):
        x = random.uniform(-5,5)
        y = random.uniform(-5,5)
        population.append((x,y))
    return population

def initialize_population4(n):
    population = []
    for i in range(n):
        x = random.uniform(-3,3)
        y = random.uniform(-3,3)
        population.append((x,y))
    return population

#The employed bee phase, where the bees update the position with a random number in the neighborhood of the current position multiplied by a random position of a bee
def employed_bee_phase1(population):
    new_population = []
    for i in range(len(population)):
        x = population[i]
        new_x = x + random.uniform(-1,1)*(x- random.choice(population))
        new_population.append(new_x)
    return new_population

def employed_bee_phase2(population):
    new_population = []
    for i in range(len(population)):
        x = population[i]
        new_x = x + random.uniform(-1,1)*(x- random.choice(population)) 
        new_population.append(new_x)
    return new_population

def employed_bee_phase3(population):
    new_population = []
    for i in range(len(population)):
        x,y = population[i]
        new_x = x + random.uniform(-1,1)*(x- random.choice(population)[0]) 
        new_y = y + random.uniform(-1,1)*(y- random.choice(population)[1]) 
        new_population.append((new_x,new_y))
    return new_population

def employed_bee_phase4(population):
    new_population = []
    for i in range(len(population)):
        x,y = population[i]
        new_x = x + random.uniform(-1,1)*(x- random.choice(population)[0]) 
        new_y = y + random.uniform(-1,1)*(y- random.choice(population)[1]) 
        new_population.append((new_x,new_y))
    return

#The onlooker bee phase, where the bees watch the dances of employed bees and choose a food source depending on the dances
def onlooker_bee_phase1(population):
    new_population = []
    fitness_values = [fitness1(x) for x in population]
    total_fitness = sum(fitness_values)
    probabilities = [f/total_fitness for f in fitness_values]
    for i in range(len(population)):
        x = random.choices(population, probabilities)[0]
        new_population.append(x)
    return new_population

def onlooker_bee_phase2(population):
    new_population = []
    fitness_values = [fitness2(x) for x in population]
    total_fitness = sum(fitness_values)
    probabilities = [f/total_fitness for f in fitness_values]
    for i in range(len(population)):
        x = random.choices(population, probabilities)[0]
        new_population.append(x)
    return new_population

def onlooker_bee_phase3(population):
    new_population = []
    fitness_values = [fitness3(x,y) for x,y in population]
    total_fitness = sum(fitness_values)
    probabilities = [f/total_fitness for f in fitness_values]
    #Choose a food source depending on the dances of employed bees
    for i in range(len(population)):
        x = random.choices(population)
        new_population.append(x)
    return new_population

def onlooker_bee_phase4(population):
    new_population = []
    fitness_values = [fitness4(x,y) for x,y in population]
    total_fitness = sum(fitness_values)
    probabilities = [f/total_fitness for f in fitness_values]
    for i in range(len(population)):
        x = random.choices(population)
        new_population.append(x)
    return new_population


#The main function to run the bee algorithm

def bee_algorithm1(n, generations):
    population = initialize_population1(n)
    for i in range(generations):
        employed_bees = employed_bee_phase1(population)
        onlooker_bees = onlooker_bee_phase1(employed_bees)
        population = onlooker_bees
        #Save the best solution found so far
        if i == 0:
            best_solution = max(population, key=fitness1)
        else:
            if max(population, key=fitness1) > best_solution:
                best_solution = max(population, key=fitness1)
    return best_solution

def bee_algorithm2(n, generations):
    population = initialize_population2(n)
    for i in range(generations):
        employed_bees = employed_bee_phase2(population)
        onlooker_bees = onlooker_bee_phase2(employed_bees)
        #The total population is the union of the employed, onlooker, and scout bees
        population = onlooker_bees
        if i == 0:
            best_solution = max(population, key=fitness2)
        else:
            if max(population, key=fitness2) > best_solution:
                best_solution = max(population, key=fitness2)
    return best_solution

def bee_algorithm3(n, generations):
    population = initialize_population3(n)
    for i in range(generations):
        employed_bees = employed_bee_phase3(population)
        onlooker_bees = onlooker_bee_phase3(employed_bees)
        population = onlooker_bees
    return population

def bee_algorithm4(n, generations):
    population = initialize_population4(n)
    for i in range(generations):
        employed_bees = employed_bee_phase4(population)
        onlooker_bees = onlooker_bee_phase4(employed_bees)
        population = onlooker_bees
    return population

#Run the bee algorithm
print(bee_algorithm1(100, 100))
print(bee_algorithm2(100, 100))
print(bee_algorithm3(100, 100))
print(bee_algorithm4(100, 100))
```
