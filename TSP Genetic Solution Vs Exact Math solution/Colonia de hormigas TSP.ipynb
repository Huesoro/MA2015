```python
import pandas as pd
import numpy as np
import random
```

```python
ants = 5  # NÃºmero de hormigas
epochs = 100  # NÃºmero de iteraciones
alpha = 1.0  # Importancia de las feromonas
beta = 1.0  # Importancia de la heurÃ­stica
evaporation_rate = 0.1  # Tasa de evaporaciÃ³n de las feromonas 
pheromone_init = 1  # Valor inicial de feromonas
lr = 1  # Tasa de aprendizaje
```

```python
df = pd.read_csv('sample0.csv')
```

```python
df = pd.read_csv('dist_matrix0.csv', index_col=0)
df = df.reset_index(drop=False)
# df = df.iloc[:50,:51]
```

```python
actions = pd.melt(df, id_vars = ['index'], var_name='Target', value_name='Cost')
actions.columns = ['origin', 'target', 'cost']
actions[['origin', 'target', 'cost']] = actions[['origin', 'target', 'cost']].apply(pd.to_numeric)
actions = actions[actions['cost'] != 999]
actions.loc[actions['cost'] == 0, 'cost'] = 0.01
actions.sort_values(by=['origin', 'target'], inplace=True)

####################

actions['pheromone_level'] = pheromone_init
actions['convenience'] = 1 / actions['cost']

actions['proportion'] = (actions['convenience'] ** beta) * (actions['pheromone_level'] ** alpha)
```

```python
for epoch in range(epochs):
    paths = []
    for ant in range(ants):
        temp_actions = actions.copy()
        origin = 1
        path = []
        path_cost = 0

        while True:
            available_targets = temp_actions.loc[temp_actions['origin'] == origin, 'target'].values
            if len(available_targets) == 0:
                break
            probabilities = temp_actions.loc[temp_actions['origin'] == origin, 'proportion'].values
            random_target = random.choices(available_targets, weights=probabilities)[0]

            path.append((origin, random_target))
            path_cost += temp_actions.loc[(temp_actions['origin'] == origin) & (temp_actions['target'] == random_target), 'cost'].values[0]

            temp_actions = temp_actions[temp_actions['target'] != random_target]
            origin = random_target

        paths.append([path, path_cost])
        actions['pheromone_level'] *= (1 - evaporation_rate) # Evaporamos las feromona
        for path, path_cost in paths:
            for origin, target in path:
                actions.loc[(actions['origin'] == origin) & (actions['target'] == target), 'pheromone_level'] += lr / path_cost
        actions['proportion'] = (actions['convenience'] ** beta) * (actions['pheromone_level'] ** alpha)

    best_path = min(paths, key=lambda x: x[1])
    print(f'Epoch {epoch + 1}: {best_path}')
    print(f'Best path: {best_path[0]}')
    print(f'Best cost: {best_path[1]}')
    print("---------------------------------")
```
